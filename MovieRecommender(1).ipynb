```python
movies=movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.dropna(inplace=True)
```

```python
movies=movies.drop_duplicates()
```

```python
import ast

def convert(obj):
    # Check if obj is already a list
    if isinstance(obj, list):
        return obj  # Return as-is if it's already a list
    try:
        # Safely evaluate the string into a Python object
        parsed = ast.literal_eval(obj)
        return [i['name'] for i in parsed]
    except (ValueError, SyntaxError, TypeError):
        # Handle any malformed or unexpected data gracefully
        return []  # Return an empty li
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    counter=0
    L=[]
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def convert4(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L        
```

```python
movies['crew']=movies['crew'].apply(convert4)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
moviess=movies[['id', 'tags', 'title']]
```

```python
moviess['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
moviess['tags']=moviess['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
moviess['tags']=moviess['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(moviess['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_ind=moviess[moviess['title']==movie].index[0]
    distances=similarity[movie_ind]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=(lambda x:x[1]))[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```
